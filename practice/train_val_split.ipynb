```python
import os
```

## 필요한 이미지만 사용 (CFD, Volker, Sylvie_Chambon, noncrack)

```python
image_path = "./crack_segmentation_dataset/images/"
mask_path = "./crack_segmentation_dataset/masks/"

image_list = os.listdir(image_path)
mask_list = os.listdir(mask_path)
```

```python
# 이름이 CFD, Volker, Sylvie, noncrack으로 시작하는 파일만 남긴다.
name = ("CFD", "Volker", "Sylvie", "noncrack")

for file_name in image_list:
    if file_name.startswith(name):
        continue
    else:
        os.remove(image_path+file_name)
        print(f"{file_name} 삭제 완료")
```

```python
for file_name in mask_list:
    if file_name.startswith(name):
        continue
    else:
        os.remove(mask_path+file_name)
        print(f"{file_name} 삭제 완료")
```

```python
## crack - noncrack으로 진행하기위해 이름 변경
image_path = "./crack_segmentation_dataset/images/"
mask_path = "./crack_segmentation_dataset/masks/"

image_list = os.listdir(image_path)
mask_list = os.listdir(mask_path)

for i in range(len(image_list)):
    if image_list[i].startswith("non"):
        continue
    else:
        os.rename(image_path+image_list[i], image_path+"crack_"+str(i)+".jpg")

for i in range(len(mask_list)):
    if mask_list[i].startswith("non"):
        continue
    else:
        os.rename(mask_path+mask_list[i], mask_path+"crack_"+str(i)+".jpg")
    
```

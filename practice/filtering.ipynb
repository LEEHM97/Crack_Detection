```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
mask_image = cv2.imread('6192.bmp')
plt.imshow(mask_image)
```

```python
real_image = cv2.imread('6192.jpg')
plt.imshow(real_image)
```

```python
#dilate 적용
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT,ksize=(3,3))
dilate = cv2.dilate(mask_image,kernel,iterations=1)

plt.imshow(dilate, cmap="gray")
```

```python
    # sigma 값을 이용하여 가우시안 필터링 적용
dst = cv2.GaussianBlur(dilate, (0, 0), 2)

plt.imshow(dst, cmap="gray")
```

```python
# 원본이미지에 가우시안 필터 적용해보기

dst = cv2.GaussianBlur(real_image, (0, 0), 3)

plt.figure(figsize=(12,12))
plt.subplot(121)
plt.imshow(dst, cmap="gray")
plt.subplot(122)
plt.imshow(real_image)
```
